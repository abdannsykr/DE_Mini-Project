```python
with open('population_data.xml', 'r') as file:
    xml_data = file.read()

# Parsing XML data with html.parser
soup = BeautifulSoup(xml_data, 'html.parser')

country_or_area = []
item = []
year = []
value = []

# Iterating through each record element
for record in soup.find_all('record'):
    # Extracting data from each field
    country_or_area.append(record.find('field', {'name': 'Country or Area'}).text)
    item.append(record.find('field', {'name': 'Item'}).text)
    year.append(record.find('field', {'name': 'Year'}).text)
    value.append(record.find('field', {'name': 'Value'}).text)

# Creating DataFrame
data_population_xml = pd.DataFrame({
    'Country_or_Area': country_or_area,
    'Item': item,
    'Year': year,
    'Value': value
})

print(data_population_xml)
```

## Ekstrak population_data.db

```python
conn = sqlite3.connect('population_data.db')
data_population_db = pd.read_sql_query("SELECT * FROM population_data", conn)
conn.close()

print(data_population_db)
```

## Cleaning population_data.db

```python
data_population_db["Indicator_Name"] = pd.to_numeric(data_population_db["Indicator_Name"], errors="coerce")

numeric_columns = data_population_db.columns[5:]

data_population_db[numeric_columns] = data_population_db[numeric_columns].fillna(data_population_db[numeric_columns].mean())
print(data_population_db.columns)
```

```python
data_population_db_duplicate = data_population_db.drop_duplicates(inplace=True)

print(data_population_db_duplicate)
```

## Ekstrak data mystery

```python
import pandas as pd

data = []
with open('mystery.csv', 'r', encoding='latin1') as file:
    for line in file:
        data.append(line.strip().split(',')) 

data_mystery = pd.DataFrame(data)
data_mystery
```

```python
data_mystery_clean = data_mystery.dropna()

print('DataFrame setelah mengatasi data yang hilang\n')
print(data_mystery_clean)
```

```python
# Menghapus duplikasi berdasarkan semua kolom kecuali kolom pertama (indeks 0)
data_mystery_duplicate = data_mystery.drop_duplicates(subset=data_mystery.columns[1:], keep='first')

# Menampilkan DataFrame setelah mengatasi duplikasi
print('DataFrame setelah mengatasi duplikasi\n')
print(data_mystery_duplicate)
```
